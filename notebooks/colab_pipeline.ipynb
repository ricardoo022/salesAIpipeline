```
import getpass

anthropic_key = getpass.getpass("ANTHROPIC_API_KEY: ")
hf_token = getpass.getpass("HF_TOKEN: ")

with open(".env", "w") as f:
    f.write(f"ANTHROPIC_API_KEY={anthropic_key}\nHF_TOKEN={hf_token}\n")

print("Wrote .env")
```

```
import os
import shutil
from google.colab import files

os.makedirs("input", exist_ok=True)
print("Select meeting.mp4 to upload:")
uploaded = files.upload()
fname = next(iter(uploaded))
shutil.move(fname, "input/meeting.mp4")
print("Saved to input/meeting.mp4")
```
